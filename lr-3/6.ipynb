```python
# частотная манипуляция цифрового сигнала

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

N = 100  # задаем длину каждого символа (количество точек выборки на символ)
np.random.seed(1)  # устанавливаем фиксированный seed для воспроизводимости случайных чисел
mod_rnd = np.random.randint(0, 2, 20)  # генерируем случайную последовательность из 0 и 1 длиной 20 элементов (битовая последовательность)
mod_fsk = np.repeat(mod_rnd, repeats=N)  # расширяем битовую последовательность, повторяя каждый бит N раз для формирования сигнала
M = mod_fsk.size  # получаем общий размер расширенного сигнала
# создаем массив частот, соответствующих битам: 10 Гц для 0 и 50 Гц для 1
mod_frq = np.zeros(M)  # инициализация массива частот нулями
mod_frq[mod_fsk == 0] = 10  # для битов равных 0 присваиваем частоту 10 Гц
mod_frq[mod_fsk == 1] = 50  # для битов равных 1 присваиваем частоту 50 Гц
# создаем FSK-сигнал, меняя частоту в зависимости от бита
sig_fsk = np.sin(mod_frq * 2.0 * np.pi * np.linspace(0, 1, M))

# создаем фигуру для отображения
plt.figure(figsize=(12, 6), dpi=80)
# первый подграфик: цифровой битовый сигнал
plt.subplot(2, 1, 1)
plt.title('Цифровой сигнал')
plt.plot(mod_fsk)  # отображаем расширенную битовую последовательность
plt.xlim([0, M-1])  # ограничиваем по оси X
plt.grid(True)  # включаем сетку

# второй подграфик: FSK-сигнал
plt.subplot(2, 1, 2)
plt.title('ЧМн-сигнал')
plt.plot(mod_fsk, '--b', linewidth=2.0)  # пунктирная синяя линия — битовая последовательность
plt.plot(sig_fsk, '-r')  # красная линия — FSK-сигнал
plt.xlim([0, M-1])  # ограничение по оси X
plt.grid(True)  # включение сетки

plt.tight_layout()  # автоматическая настройка расположения для лучшей визуализации
```
