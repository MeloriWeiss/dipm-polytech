```python
# фазовая манипуляция цифрового сигнала

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

N = 200  # задаем длину каждого символа (количество точек выборки на символ)
np.random.seed(1)  # фиксируем зерно генератора случайных чисел для воспроизводимости
mod_rnd = np.random.randint(0, 2, 25)  # создаем случайную последовательность из 0 и 1 длиной 25 элементов (битовая последовательность)
mod_psk = np.repeat(mod_rnd, repeats=N)  # расширяем последовательность, повторяя каждый бит N раз для формирования сигнала
M = mod_psk.size  # определяем общий размер расширенного сигнала
# создаем фазовый сигнал для ФМ-модуляции
# частота 25 Гц, добавляем фазовый сдвиг, зависящий от битов (1 или 0)
sig_psk = np.sin(25 * 2.0 * np.pi * np.linspace(0, 1, M) + np.pi * mod_psk)

# создаем фигуру для отображения
plt.figure(figsize=(12, 6), dpi=80)
# первый подграфик: цифровой битовый сигнал
plt.subplot(2, 1, 1)
plt.title('Цифровой сигнал')
plt.plot(mod_psk, color='C0', linewidth=2.0)  # отображаем расширенную последовательность бит
plt.xlim([0, M-1])  # ограничиваем по оси X
plt.grid(True)  # включаем сетку

# второй подграфик: ФМ-сигнал
plt.subplot(2, 1, 2)
plt.title('ФМн-сигнал')
plt.plot(mod_psk, '--b', linewidth=2.0)  # пунктирная синяя линия — битовая последовательность
plt.plot(sig_psk, '-r')  # красная линия — сигнал с частотной модуляцией
plt.xlim([0, M-1])  # ограничение по оси X
plt.grid(True)  # включение сетки

plt.tight_layout()  # автоматическая настройка расположения для лучшей визуализации
```
