```python
# построение графиков несущего, модулирующего и ЧМ-сигнала

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# функция для генерации частотно-модулированного (ЧМ) сигнала
def signal_fm(amp=1.0, kd=0.25, fc=10.0, fs=2.0, period=100):
    # создаем массив времени t, равномерно распределенный от 0 до 2π
    t = 2.0 * np.pi * np.linspace(0, 1, period)
    # возвращаем значение ЧМ сигнала:
    # - амплитуда: 'amp'
    # - аргумент косинуса включает базовую частоту 'fc' и модуляцию через функцию синуса
    return amp * np.cos(fc * t + (kd / fs) * np.sin(fs * t))

N = 1024  # задаем длину сигнала (количество точек выборки)
sig = signal_fm(amp=1.0, kd=15, fc=40, fs=4, period=N)  # генерируем частотно-модулированный сигнал с заданными параметрами
smd = np.cos(4 * 2.0 * np.pi * np.linspace(0, 1, N))  # создаем модулирующий сигнал (косинус с частотой 4 Гц)
car = np.sin(40 * 2.0 * np.pi * np.linspace(0, 1, N))  # создаем несущий сигнал (синус с частотой 40 Гц)
plt.figure(figsize=(12, 7), dpi=80)  # создаем графическую фигуру размером 12x7 дюймов с разрешением 80 точек на дюйм
plt.subplot(2, 1, 1)  # создаем первый подграфик в сетке 2x1 (верхний)
plt.plot(smd, label='Модулирующий информационный сигнал')  # строим модулирующий сигнал на первом подграфике
plt.plot(car, label='Несущий сигнал')  # строим несущий сигнал на том же графике
plt.xlim([0, N//2 - 1])  # устанавливаем лимит по оси X от 0 до половины длины сигнала (N/2 - 1)
plt.grid(True)  # включаем сетку для удобства чтения графика
plt.legend()  # добавляем легенду для обозначения сигналов
plt.subplot(2, 1, 2)  # создаем второй подграфик в сетке 2x1 (нижний)
plt.title('Частотно-модулированный сигнал')  # задаем заголовок для второго графика
plt.plot(sig)  # строим частотно-модулированный сигнал
plt.xlim([0, N//2 - 1])  # устанавливаем лимит по оси X от 0 до половины длины сигнала
plt.grid(True)  # включаем сетку для удобства чтения графика
```
