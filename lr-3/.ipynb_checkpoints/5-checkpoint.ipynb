```python
# амплитудная манипуляция цифрового сигнала

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

N = 64  # задаем длину одного символа (количество точек выборки на символ)
np.random.seed(1)  # устанавливаем фиксированныйseed для воспроизводимости случайных чисел
mod_rnd = np.random.randint(0, 2, 40)  # генерируем случайную последовательность из 0 и 1 длиной 40 элементов (битовая последовательность)
mod_ask = np.repeat(mod_rnd, repeats=N)  # расширяем битовую последовательность, повторяя каждый бит N раз для формирования сигнала
M = mod_ask.size  # получаем общий размер расширенного сигнала
# создаем амплитудно-импульсную модуляцию: бит умножается на синусоиду с частотой 64 Гц
sig_ask = mod_ask * np.sin(64 * 2.0 * np.pi * np.linspace(0, 1, M))
plt.figure(figsize=(12, 6), dpi=80)  # создаем фигуру для графиков

plt.subplot(2, 1, 1)  # первый подграфик (верхний)
plt.title('Цифровой сигнал')  # заголовок для цифрового сигнала
plt.plot(mod_ask)  # строим расширенную битовую последовательность
plt.xlim([0, M-1])  # ограничение по оси X от 0 до M-1
plt.grid(True)  # включаем сетку

plt.subplot(2, 1, 2)  # второй подграфик (нижний)
plt.title('АМн-сигнал')  # заголовок для амплитудно-импульсного модулированного сигнала
plt.plot(mod_ask, '--b', linewidth='2.0')  # строим линию с пунктирной синей линией, отображающую битовую последовательность
plt.plot(sig_ask, 'r')  # строим амплитудно-импульсный сигнал красным цветом
plt.xlim([0, M-1])  # ограничение по оси X
plt.grid(True)  # включаем сетку

plt.tight_layout()  # автоматически настраиваем расположение графиков для лучшей видимости
```
