```python
# вывод графиков

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# создаем равномерный массив времени от 0 до 1 с 100 точками
t = np.linspace(0, 1, 100, endpoint=True)

# создаем сигнал — косинус с частотой 1 Гц (2*pi*t)
f = np.cos(2*np.pi*t)

# задаем количество отсчётов для дискретизации
steps = 20

# создаем фигуру для графика с заданным размером
fig = plt.figure(figsize=(6, 3))

# создаем массив времени для дискретных точек — равномерно распределенных по интервалу [0, 1]
tt = np.linspace(0, 1, steps, endpoint=True)

# строим исходный сигнал как линию (синий цвет, сплошная линия)
plt.plot(t, f, '-b')

# строим ступенчатый график для дискретных точек в виде зеленых ступенек
plt.step(tt, np.cos(2*np.pi*tt), 'g')

# добавляем красные пунктирные линии, соединяющие дискретные точки
plt.plot(tt, np.cos(2*np.pi*tt), '--r')

# включаем сетку для удобства восприятия
plt.grid()

# ограничиваем область по оси x от 0 до 1
plt.xlim([0, 1])

# автоматически подбираем расположение элементов так, чтобы не было наложений
plt.tight_layout()
```
