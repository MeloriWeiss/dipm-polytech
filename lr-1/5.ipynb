```python
# быстрое преобразование Фурье

# импортируем необходимые библиотеки
import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# параметры сигнала
amplitude = 1.0  # Амплитуда гармонического сигнала
frequency = 100.0  # Частота гармонического сигнала (Гц)
duration = 1.0  # Длительность сигнала в секундах
sampling_rate = 1000  # Частота дискретизации (количество отсчетов в секунду)

# создаем временную ось от 0 до duration, с шагом, определенным sampling_rate
t = np.linspace(0, duration, int(duration * sampling_rate), endpoint=False)

# создаем гармонический сигнал: синусоида с заданной частотой
signal = amplitude * np.sin(2 * np.pi * frequency * t)

# вычисляем быстрое преобразование Фурье (FFT) сигнала
fft = np.fft.fft(signal)

# вычисляем соответствующие частоты для каждого компонента FFT
frequencies = np.fft.fftfreq(len(signal), 1 / sampling_rate)

# строим график спектра: абсолютное значение FFT по частотам
plt.figure()
plt.plot(frequencies, np.abs(fft))

# ограничиваем график по частоте до половины частоты дискретизации (независимо от спектра)
plt.xlim(0, sampling_rate/2)

# добавляем подписи и заголовок
plt.xlabel('Частота (Гц)')
plt.ylabel('Амплитуда')
plt.title('Спектр гармонического сигнала')

# включаем сетку для удобства чтения
plt.grid(True)
```
