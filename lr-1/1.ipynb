```python
# вывод графиков

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# создаем вектор времени t от 0 до 1 с 50 точками
t = np.linspace(0, 1, 50, endpoint=True)

# создаем сигнал f как сумму нескольких косинусоидальных волн с разными частотами и фазами
f = 3*np.cos(np.pi*t) + np.cos(3*np.pi*t) + np.cos(0.5*np.pi*t) + np.cos(7*np.pi*t+5)

# названия трех графиков, которые мы будем отображать
titles = ['Аналоговый сигнал', 'Дискретный сигнал', 'Квантованный сигнал']

# создаем фигуру для отображения трех графиков, задаем размер 16 на 4 дюйма
fig = plt.figure(figsize=(16, 4))

# цикл по трем графикам (0, 1, 2)
for i in range(3):
    # создаем подграфик (1 строка, 3 столбца, i+1-я позиция)
    plt.subplot(1, 3, i+1)

    # устанавливаем название текущего графика
    plt.title(titles[i])
    
    # в зависимости от номера графика, используем разные типы отображения
    if i == 0:
        # первый график: линийный (сплошная линия)
        plt.plot(t, f)
    if i == 1:
        # второй график: дискретный (столбчатая или точечная), используем stem — "стерженьки"
        plt.stem(t, f)
    if i == 2:
        # третий график: ступенчатая линия для отображения дискретных уровней
        plt.step(t, f)
    
    # ограничиваем область по оси X от 0 до 1
    plt.xlim([0, 1])
    
    # устанавливаем метки по оси Y с равными интервалами от 6 до -6 (13 точек)
    plt.yticks(np.linspace(6, -6, 13))
    
    # включаем сетку для удобства восприятия графика
    plt.grid(True)
    
    # получаем текущие оси
    ax = plt.gca()
    # удаляем правую и верхнюю рамки графика для более аккуратного вида
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    # смещаем нижнюю рамку (ось X) так, чтобы она проходила через ноль по данным
    ax.spines['bottom'].set_position(('data', 0))

# подбираем расположение графиков так, чтобы они не перекрывались
plt.tight_layout()
```
