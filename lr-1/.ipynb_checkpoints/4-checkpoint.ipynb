```python
# задание и вывод графика функции Хевисайда

# импортируем необходимые библиотеки
import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# задаем параметр n — половина диапазона по времени
n = 6

# создаем вектор времени t от -n до n-1, всего 2*n точек
t = np.linspace(-n, n-1, 2*n)

# вычисляем функцию Хевисайда для каждого значения t
# второй аргумент 1 означает, что функция равна 1 при t >= 0, и 0 при t < 0
fh = np.heaviside(t, 1)

# создаем фигуру с заданным размером и разрешением
fig = plt.figure(figsize=(6, 3), dpi=100)

# строим стебельчатую диаграмму для функции Хевисайда
# маркер D — ромб, отображающий точки
plt.stem(t, fh, markerfmt='D')

# добавляем подпись по оси x
plt.xlabel('Отсчёты')

# задаем метки по оси x для каждого элемента t
plt.xticks(t)

# ограничиваем диапазон по оси x для лучшей видимости
plt.xlim([np.min(t)+1, np.max(t)])

# включаем сетку
plt.grid(True)

# автоматически подбираем расположение элементов графика
plt.tight_layout()
```
