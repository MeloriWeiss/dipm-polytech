```python
# задание и вывод графика дельта-функции

# импортируем необходимые библиотеки
import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# задаем параметр n — половина длины вектора времени (от -n до n-1)
n = 6

# создаем вектор времени t от -n до n-1, всего 2*n точек
t = np.linspace(-n, n-1, 2*n)

# создаем массив fd нулей той же длины, что и t
fd = np.zeros(2*n)

# устанавливаем значение 1 в центральную точку (имитируя дельта-функцию)
fd[n] = 1

# создаем фигуру с заданным размером и разрешением (dpi)
fig = plt.figure(figsize=(6, 3), dpi=100)

# строим стебельчатую диаграмму (stem plot) для дельта-функции
# маркер D — ромб, отображающий точку
plt.stem(t, fd, markerfmt='D')

# добавляем подпись по оси x
plt.xlabel('Отсчёты')

# задаем метки по оси x для каждого элемента t
plt.xticks(t)

# ограничиваем диапазон по оси x, чтобы было видно все точки с небольшим отступом
plt.xlim([np.min(t)+1, np.max(t)])

# включаем сетку для лучшей читаемости
plt.grid(True)

# автоматически подбираем расположение элементов графика, чтобы избежать наложений
plt.tight_layout()
```
