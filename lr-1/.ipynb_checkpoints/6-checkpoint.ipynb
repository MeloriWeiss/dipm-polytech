```python
# построение амплитудного спектра

# импортируем необходимые библиотеки
import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# задаем параметры: длительность сигнала и частоту дискретизации
duration = 1.0  # длительность сигнала в секундах
sampling_rate = 1000  # частота дискретизации в герцах
# создаем временную ось
t = np.linspace(0, duration, int(duration * sampling_rate), endpoint=False)

# создаем два гармонических сигнала с разными частотами
signal1 = np.sin(2 * np.pi * 10 * t)   # сигнал с частотой 10 Гц
signal2 = np.sin(2 * np.pi * 100 * t)  # сигнал с частотой 100 Гц
# создаем сумму двух сигналов
signal3 = signal1 + signal2

# вычисляем быстрое преобразование Фурье для каждого сигнала
fft1 = np.fft.fft(signal1)
fft2 = np.fft.fft(signal2)
fft3 = np.fft.fft(signal3)

# получаем массив частот, соответствующих компонентам FFT
freqs = np.fft.fftfreq(len(t), 1 / sampling_rate)

# построение спектров для signal1 и signal2 в одной фигуре
plt.figure(figsize=(10, 4))
# спектр сигнала 1
plt.subplot(1, 2, 1)
plt.plot(freqs, np.abs(fft1))
plt.title('спектр сигнала 1 (10 гц)')
# ограничиваем ось по частоте для лучшей читаемости
plt.xlim(0, 200)

# спектр сигнала 2
plt.subplot(1, 2, 2)
plt.plot(freqs, np.abs(fft2))
plt.title('спектр сигнала 2 (100 гц)')
plt.xlim(0, 200)

plt.tight_layout()

# построение спектра суммы сигналов в отдельной фигуре
plt.figure(figsize=(5, 4))
plt.plot(freqs, np.abs(fft3))
plt.title('спектр суммы сигналов')
plt.xlim(0, 200)

plt.tight_layout()
```
