```python
# вывод сигнала и его амплитудного спектра

# импортируем необходимые библиотеки
import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# задаем размер сигнала
N = 1024
# создаем исходный прямоугольный сигнал
x = np.zeros(N)
x[128:320] = 1  # устанавливаем интервал с значением 1

# находим бпф исходного сигнала
X = np.fft.fft(x, N)
# сдвигаем нулевую частоту в центр спектра и берем модуль (амплитуду)
Xs = np.fft.fftshift(np.abs(X))
# создаем массив нормализованных частот от -0.5 до 0.5
f = np.linspace(-0.5, 0.5, N, endpoint=True)

# создаем фигуру для отображения
plt.figure(figsize=(14, 3), dpi=100)

# выводим график исходного сигнала
plt.subplot(1, 2, 1)
plt.plot(x)
plt.title('прямоугольный сигнал')
plt.xlabel('отсчёты')
plt.ylabel('')
plt.xlim([0, N-1])
# задаем метки по оси x с равными интервалами
plt.xticks(np.linspace(0, N, 9, endpoint=True))
plt.grid()

# выводим амплитудный спектр
plt.subplot(1, 2, 2)
plt.stem(f, Xs)
plt.title('амплитудный спектр')
plt.xlabel('частота')
plt.ylabel('амплитуда')
# ограничиваем ось по частоте для выделения основной полосы
plt.xlim([-1/16, 1/16])
# задаем метки по оси x
plt.xticks(np.linspace(-1/16, 1/16, 6, endpoint=True))
plt.grid()

plt.tight_layout()
```
