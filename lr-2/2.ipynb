```python
# нахождение значения сигнал при помощи δ-функции

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# задаем размер массива N равным 10
N = 10
# создаем дельта-функцию: массив из нулей длиной N
fd = np.zeros(N)
# устанавливаем значение 1 в центральную точку массива (имитируя дельта-функцию)
fd[5] = 1
# устанавливаем зерно генератора случайных чисел для воспроизводимости
np.random.seed(2)
# генерируем случайный сигнал из N нормальных случайных чисел
x = np.random.randn(N)
# вычисляем кросс-корреляцию между дельта-функцией и случайным сигналом
r = np.correlate(fd, x)
# создаем график с заданными размерами и разрешением
plt.figure(figsize=(14, 4), dpi=80)
# строим график дельта-функции с маркерами
plt.plot(fd, '-o')
# строим график случайного сигнала с маркерами
plt.plot(x, '-s')
# ограничиваем ось x от -0.5 до N-0.5 для четкого отображения точек
plt.xlim([-0.5, N-0.5])
# включаем отображение сетки на графике
plt.grid(True)
# выводим значение кросс-корреляции и значение сигнала в позиции 5
print('r =', r, ' x[5] =', x[5])
```
