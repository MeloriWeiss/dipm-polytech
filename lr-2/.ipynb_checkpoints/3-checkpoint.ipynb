```python
# вычисление линейной свёртки

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

N, M = 4, 3
a = [1, 2, 3, 4]
b = [3, 2, 1]
# Дополняем нулями слева первый сигнал до длины N+M-1
an = np.concatenate([np.zeros(M-1, dtype=int), a])
# Инвертируем во времени второй сигнал и дополняем его нулями справа N+M-1
bn = np.concatenate([b[::-1], np.zeros(N-1, dtype=int)])
print('an =', an)
print('bn =', bn)
ab = []
for i in range(N+M-1):
    # сдвигаем второй сигнал вправо
    br = np.roll(bn, i)
    # вычисляем произведения элементов и подсчитываем сумму произведений
    sm = np.sum(an * br)
    # добавляем значение в список
    ab.append(sm)
print('\na * b =', ab)

# ------ сравнение ------
# вычисляем свёртку
cv = np.convolve(a,b, mode='full')
print('np.convolve =', cv)
# Сравниваем значения
ab_check = np.all(ab == cv)
print(ab_check)
```
