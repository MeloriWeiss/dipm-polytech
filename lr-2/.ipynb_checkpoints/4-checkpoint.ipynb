```python
# вычисление циклической свёртки

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

N = 4
a = np.array([1, 2, 3, 4], dtype=int)
b = np.array([3, 2, 1, 0], dtype=int)
# инвертируем второй сигнал
bi = b[::-1]
ab = []
for i in range(N):
    # сдвигаем второй сигнал
    br = np.roll(bi, i+1)
    # вычисляем произведения элементов и подсчитываем сумму произведений
    sm = np.sum(a * br)
    # добавляем значение в список
    ab.append(sm)
print('a * b =', ab)


# вычисление циклической свёртки с помощью БПФ
# Вычисляем циклическую свёртку
cv = np.real(np.fft.ifft(np.fft.fft(a) * np.fft.fft(b)))
print('circular convolution = ', cv)
ab_check = np.all(ab == cv)
print(ab_check)
```
