```python
# случайные процессы с нулевой корреляцией

import numpy as np
import matplotlib.pyplot as plt

# команда для отображения графиков прямо внутри ноутбука
%matplotlib inline

# нахождение корреляции сигнала с самим собой
# x1 = np.array([1, 2, 3, 4, 5, 6, 7])
# np.correlate(x1, x1)

# нахождение корреляции по формуле
# np.sum(x1*x1)

# отрицательная корреляция
# x1 = np.array([1, 2, 3])
# x2 = np.array([-1, -2, -3])
# np.correlate(x1, x2)

# задаем размер массива N равным 20
N = 20
# устанавливаем зерно генератора случайных чисел для воспроизводимости
np.random.seed(1)
# генерируем массив x1 из 20 случайных целых чисел в диапазоне от -10 до 9
x1 = np.random.randint(-10, 10, N)
# генерируем второй массив x2 аналогично из 20 случайных целых чисел
x2 = np.random.randint(-10, 10, N)
# вычисляем кросс-корреляцию между x1 и x2 с помощью функции np.correlate
r12 = np.correlate(x1, x2)
# создаем график с заданными размерами (ширина 14, высота 4) и разрешением 80 точек на дюйм
plt.figure(figsize=(14, 4), dpi=80)
# строим график массива x1 с маркерами и линиями
plt.plot(x1, '-d')
# строим график массива x2 с маркерами и линиями
plt.plot(x2, '-p')
# ограничиваем ось x от -0.5 до N-0.5 для четкого отображения точек
plt.xlim([-0.5, N-0.5])
# включаем отображение сетки на графике
plt.grid(True)
# выводим значение кросс-корреляции
print('r12 =', r12)
```
